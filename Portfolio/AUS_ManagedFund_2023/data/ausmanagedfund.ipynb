```python
import pandas as pd
import os
import json

cwd = os.getcwd()
fundSummary = pd.read_excel(cwd + '/Summary_Managed Fund_Institutions.xlsx', sheet_name='Data1')
```

```python
df = fundSummary.iloc[9:,:]
```

```python
df1 = df.iloc[-8:,1:13]
df1
```

```python
df1_pct = df1.copy()
df1_pct.iloc[:,1:] = df1_pct.iloc[:,1:].apply(lambda x: x/x.sum()*100, axis=1)
df1_pct
```

```python
df1_pct_sort = df1_pct.iloc[:,[0,11,7,6,9,1]]
df1_pct_sort['Other'] = df1_pct.iloc[:,[8,3,2,4,10,5]].apply(lambda x: x.sum(), axis=1)
```

```python
## Data for chart 1
df1_pct_sort["Period"] = ['Q1_2022','Q2_2022','Q3_2022','Q4_2022','Q1_2023','Q2_2023','Q3_2023','Q4_2023']        
df1_pct_sort = df1_pct_sort.iloc[:,[-1,0,1,2,3,4,5,6]]
df1_pct_sort
```

```python
df2 = df.iloc[-9:,20:].reset_index(drop=True) 
```

```python
df2_growth = df2.copy()
for i, d in df2_growth.iterrows():
    if i == 0:
        df2_growth.loc[i,:] = 0
    else:
        df2_growth.loc[i,:] = (df2.loc[i,:] - df2.loc[i-1,:])/df2.loc[i-1,:]*100
df2_growth = df2_growth.iloc[1:,:]
df2_growth.columns = ['Total Managed Funds assets','Life insurance corporations','Superannuation (pension) funds','Public offer (retail) unit trusts','Friendly societies','Common funds','Cash management trusts']
df2_growth["Period"] = ['Q1_2022','Q2_2022','Q3_2022','Q4_2022','Q1_2023','Q2_2023','Q3_2023','Q4_2023']        
df2_growth = df2_growth.iloc[:,[7,0,1,2,3,4,5,6]]
df2_growth
```

```python
# data for chart 1
df1_pct_sort.to_json('industry_sources.json', orient='records', lines=True, indent=2)
df2_growth.to_json('institution_growth.json', orient='records', lines=True, indent=2)
```
