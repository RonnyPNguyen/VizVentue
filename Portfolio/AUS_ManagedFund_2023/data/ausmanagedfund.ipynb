```python
import pandas as pd
import numpy as np
import os
import json

cwd = os.getcwd()
fundSummary = pd.read_excel(cwd + '/Summary_Managed Fund_Institutions.xlsx', sheet_name='Data1')
```

```python
df = fundSummary.iloc[9:,:]
```

```python
df1 = df.iloc[-8:,1:13]
df1
```

```python
df1_pct = df1.copy()
df1_pct.iloc[:,1:] = df1_pct.iloc[:,1:].apply(lambda x: x/x.sum()*100, axis=1).round(2)
df1_pct
```

```python
df1_pct_sort = df1_pct.iloc[:,[0,11,7,6,9,1]]
df1_pct_sort['Other'] = df1_pct.iloc[:,[8,3,2,4,10,5]].apply(lambda x: x.sum(), axis=1)
```

```python
## Data for chart 1
df1_pct_sort["Period"] = ['Q1_2022','Q2_2022','Q3_2022','Q4_2022','Q1_2023','Q2_2023','Q3_2023','Q4_2023']        
df1_pct_sort = df1_pct_sort.iloc[:,[-1,0,1,2,3,4,5,6]]
df1_pct_sort
```

```python
df1_pct_sort.columns = ['Period','Total Assets','Overseas assets','Units in trusts','Shares','Land, buildings and equipment','Deposits','Others']
```

```python
df1_pct_sort
```

```python
df2 = df.iloc[-9:,20:].reset_index(drop=True) 
```

```python
df2_growth = df2.copy()
for i, d in df2_growth.iterrows():
    if i == 0:
        df2_growth.loc[i,:] = 0
    else:
        df2_growth.loc[i,:] = (df2.loc[i,:] - df2.loc[i-1,:])/df2.loc[i-1,:]*100
df2_growth = df2_growth.iloc[1:,:]
df2_growth = df2_growth.astype(float).round(2)
df2_growth.columns = ['Total managed funds assets','Life insurance corporations','Superannuation (pension) funds','Public offer (retail) unit trusts','Friendly societies','Common funds','Cash management trusts']
df2_growth["Period"] = ['Q1_2022','Q2_2022','Q3_2022','Q4_2022','Q1_2023','Q2_2023','Q3_2023','Q4_2023']
df2_growth = df2_growth.iloc[:,[7,0,1,2,3,4,5,6]]
df2_growth
```

```python
df2_growth_pct = pd.DataFrame(columns=['Period', 'Institution', 'Rate'])
for i, d in df2_growth.iterrows():
    rows_to_add = []
    for j in range(1, 8):
        row = {'Period': d['Period'], 'Institution': df2_growth.columns[j], 'Rate': d[j]}
        rows_to_add.append(row)
    temp_df = pd.DataFrame(rows_to_add)
    df2_growth_pct = pd.concat([df2_growth_pct, temp_df], ignore_index=True)
```

```python
def first_letters(text,skip_letter='('):
    # Split the text into words and collect the first letter of each word
    return ''.join(word[0] for word in text.split() if not word.lower().startswith(skip_letter.lower()))
df2_growth_pct['identifier'] = df2_growth_pct['Institution'].apply(first_letters)
df2_growth_pct.sample(5)
```

```python
df1_pct_sort.to_csv(cwd + '/df1_pct_sort.csv', index=False)
df2_growth_pct.to_csv(cwd + '/df2_growth_pct.csv', index=False)
```
