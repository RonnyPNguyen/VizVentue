```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

cwd = os.getcwd()
```

```python
input1 = pd.read_excel(cwd + '/home_ownership_rates.xlsx')
input1[input1["ownership_rate"] == 0] = np.nan
input1[["state", "birth_year", "age_group"]] = input1[["state", "birth_year","age_group"]].astype(str)
input1.dropna(inplace=True)
input1["generation"] = input1["birth_year"].map({
    "1947–1951": "Baby Boomers",
    "1952–1956": "Baby Boomers",
    "1957–1961": "Baby Boomers",
    "1962–1966": "Generation X",
    "1967–1971": "Generation X",
    "1972–1976": "Generation X",
    "1977–1981": "Generation X",
    "1982–1986": "Millennials",
    "1987–1991": "Millennials",
    "1992–1996": "Millennials"
})
input1
```

```python
input1.to_csv(cwd + '/home_ownership_rates.csv', index=False)
```
